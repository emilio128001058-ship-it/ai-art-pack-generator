```python
# Cell 3 — Generate pixel art images
import requests, random, os, json
from IPython.display import display, Image as IPImage
from google.colab import files

SUBJECTS = [
    "a knight in armor in a forest",
    "a wizard casting a spell",
    "a dragon breathing fire",
    "a medieval village at sunset",
    "a space explorer on an alien planet",
    "a pirate ship on the ocean",
    "a dungeon with treasure chests",
    "a cozy tavern interior",
    "a witch's potion shop",
    "a cyberpunk street at night",
    "a fairy garden with glowing flowers",
    "a robot in a neon city",
]

STYLES = [
    "16-bit pixel art, SNES style",
    "8-bit pixel art, NES style",
    "pixel art, isometric retro RPG",
    "pixel art, top-down RPG sprite style",
    "pixel art, side-scrolling platformer style",
]

os.makedirs("pixel_art", exist_ok=True)
saved = []

for i in range(1, HOW_MANY + 1):
    subject = random.choice(SUBJECTS)
    style   = random.choice(STYLES)
    prompt  = f"{subject}, {style}, vibrant colors, sharp edges, no blur, retro game aesthetic"

    print(f"[{i}/{HOW_MANY}] {subject} | {style}")

    resp = requests.post(
        API_URL,
        headers={"Authorization": f"Bearer {API_KEY}", "Accept": "image/*"},
        data={"prompt": prompt, "output_format": "png", "aspect_ratio": "1:1"},
        files={"none": ""},
        timeout=60,
    )

    if resp.ok:
        path = f"pixel_art/pixel_art_{i:02d}.png"
        with open(path, "wb") as f:
            f.write(resp.content)
        print(f"  ✓ saved {path} ({len(resp.content)//1024} KB)")
        display(IPImage(filename=path, width=300))
        saved.append(path)
    else:
        print(f"  ✗ ERROR {resp.status_code}: {resp.text[:200]}")

print(f"\nDone! {len(saved)}/{HOW_MANY} images saved. ~{len(saved)*3} credits used.")
```

```python
# Cell 4 — Download all images as a zip
import zipfile
from google.colab import files

with zipfile.ZipFile("pixel_art_pack.zip", "w") as z:
    for path in saved:
        z.write(path)

print("Downloading pixel_art_pack.zip...")
files.download("pixel_art_pack.zip")
```
